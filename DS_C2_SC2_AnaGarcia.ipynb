```python
#pip install pandas

#a. Extraer la información del archivo.
import pandas as pd
archivo_csv = "/content/country_vaccinations.csv"
data = pd.read_csv(archivo_csv)

#b. Mostrar la estructura y tipos de datos de cada columna para identificar qué
#operaciones puedes realizar con cada una de ellas, asegurándote que las
#columnas con fechas sean del tipo datetime64.
print(data.info())
data['date'] = pd.to_datetime(data['date'])
print(data.info())

#c. Determinar la cantidad de vacunas aplicadas de cada compañía
#(con base en cómo lo reporta cada país en la columna vaccines,
#en otras palabras, agrupe por vaccines y realice la sumatoria).
# Agrupar valores de la columna M y realizar la sumatoria
suma_por_grupo = data.groupby('vaccines').size().reset_index(name='Suma')
print(suma_por_grupo)

#d. Obtener la cantidad de vacunas aplicadas en todo el mundo.
suma_D = data['total_vaccinations'].sum()
print(f"Vacunas aplicadas en todo el mundo:{suma_D}")

#e. Calcular el promedio de vacunas aplicadas por país.
promedio_vacunas_por_pais = data.groupby('country')['total_vaccinations'].mean()
print(promedio_vacunas_por_pais)

#f. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.
valor_D = data[data['date'] == '29/01/21']['total_vaccinations'].values[0]
print(f"Vacunas aplicadas el día 29/01/21 en todo el mundo: {valor_D}")

#g. Crear un dataframe nuevo denominado conDiferencias que contenga los datos
#originales y una columna derivada (diferencias) con las diferencias de
#aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.
conDiferencias = pd.DataFrame()
conDiferencias['daily_vaccinations_raw'] = data['daily_vaccinations_raw']
conDiferencias['daily_vaccinations'] = data['daily_vaccinations']
conDiferencias['diferencias'] = data['daily_vaccinations_raw'] - data['daily_vaccinations']

#h. Obtener el periodo de tiempo entre el registro con fecha más reciente y el
#registro con fecha más antigua.
fecha_mas_reciente = data['date'].max()
fecha_mas_antigua = data['date'].min()
periodo_tiempo = fecha_mas_reciente - fecha_mas_antigua
print(f"Tiempo entre el registro con fecha más reciente y el registro con fecha más antigua: {periodo_tiempo}")

#i. Crear un dataframe nuevo denominado conCantidad que contenga los datos
#originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas
#cada día (usar la columna vaccines y separar por el carácter , ).
conCantidad = pd.DataFrame()
conCantidad['vaccines'] = data['vaccines']
conCantidad['canVac'] = data['vaccines'].astype(str) + ',' + data['country'].astype(str)

#j. Generar un dataframe denominado antes20 con todos los registros que se hayan
#realizado antes del 20 de diciembre de 2020.
antes20 = data[data['date'] < '12/20/2020']

#k. Obtener un dataframe denominado pfizer con todos los registros donde se haya
#utilizado la vacuna Pfizer.
pfizer = data[data['vaccines'] == 'Pfizer/BioNTech']

#l. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20
#y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada
#dataframe ocupe una hoja diferente.
with pd.ExcelWriter('resultadosReto.xlsx') as writer:
    conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=False)
    conCantidad.to_excel(writer, sheet_name='conCantidad', index=False)
    antes20.to_excel(writer, sheet_name='antes20', index=False)
    pfizer.to_excel(writer, sheet_name='pfizer', index=False)

```
